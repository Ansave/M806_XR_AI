```python
x = np.arange(0, image.shape[1], 1)
y = np.arange(0, image.shape[0], 1)

X,Y = np.meshgrid(x,y)
Z = np.mean(image, axis=2)
Z = 1 - Z
# Z = image
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
ax.invert_yaxis()
plt.axis("off")

# my_cmap = "terrain"
my_cmap = "gray"
# my_cmap = "viridis"

CS = ax.contour(X, Y, Z, levels=5, linewidths=1.5, cmap=my_cmap)
CS = ax.contour(X, Y, Z, levels=25, linewidths=0.5, cmap=my_cmap)
CS = ax.contour(X, Y, Z, levels=50, linewidths=0.2, cmap=my_cmap)

# ax.clabel(CS, inline=True, fontsize=10)
# ax.set_title('Simplest default with labels')

fig.set_size_inches(10,10)
fig.savefig("Contourmap.svg")
# fig.savefig("Contourmap.png")
```

```python
# from 0 to 1

fname = f"DepthRenders/Image000{i}.png"
image = plt.imread(fname)

f, ax = plt.subplots()
f.set_size_inches(6,6)
# plt.axis("off")

image = image[:,:,3]
image = 1 - image

levels_count = 20
CS = ax.contour(X, Y, Z + image.max() / levels_count / 2, levels=levels_count, linewidths=.5, colors="black")
CS = ax.contour(X, Y, Z + image.max() / levels_count / 2, levels=levels_count*2, linewidths=.1, colors="black")
image = np.round(image / (image.max() / levels_count)) / levels_count

# my_cmap = "gray"
my_cmap = "terrain"

im = ax.imshow(image, cmap=my_cmap)

from mpl_toolkits.axes_grid1 import make_axes_locatable
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="3%", pad=0.20)
plt.colorbar(im, cax=cax)

f.set_size_inches(10,10)
f.savefig("FilledContourMap.svg")
f.savefig("FilledContourMap.png", facecolor='white', bbox_inches="tight")
# f.savefig("FilledContourMap.pdf", bbox_inches="tight")

```

```python
# Import Library

import numpy as np 
import matplotlib.pyplot as plt 

# Define Data

x1 = [2.5, 3.5, 4.5, 5.5, 6.5, 7.5]
y1 = [3, 6, 9, 12, 15, 18]

x2 = [5, 10, 15, 20, 25, 30]
y2 = [2.6, 9.5, 14, 12, 8.3, 12]

r = np.arange(6)
width = 0.4

# Draw first subplot

plt.subplot(1, 2, 1)
plt.bar(r, x1, width=width)
plt.bar(r + width, y1, width=width)

# Draw second subplot

plt.subplot(1, 2, 2)
plt.bar(r, x2, width=width)
plt.bar(r + width, y2, width=width)

# Display

plt.show()
```
