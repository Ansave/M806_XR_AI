```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.ticker import FixedLocator
```

```python
size = np.array([2**x for x in range(15)], dtype = float)

time_1CPU = np.loadtxt("1CPU_time.txt")
time_2CPU = np.loadtxt("2CPU_time.txt")
time_4CPU = np.loadtxt("4CPU_time.txt")
time_8CPU = np.loadtxt("8CPU_time.txt")
time_16CPU = np.loadtxt("16CPU_time.txt")
```

```python
fig = plt.figure(figsize = (15, 7))
ax = plt.axes()

plt.plot(size[:], time_1CPU[:], '-o', label = "Кол-во ядер: 1")
plt.plot(size[1::], time_2CPU[:], '-o', label = "Кол-во ядер: 2")
plt.plot(size[2::], time_4CPU[:], '-o', label = "Кол-во ядер: 4")
plt.plot(size[3::], time_8CPU[:], '-o', label = "Кол-во ядер: 8")
plt.plot(size[4::], time_16CPU[:], '-o', label = "Кол-во ядер: 16")

ax.set_xscale('log', base=2)
ax.set_yscale('log', base=2)
ax.set_xlabel("Размер матрицы")
ax.set_ylabel("Время выполнения")
ax.xaxis.set_major_locator(FixedLocator([2**x for x in range(15)]))
ax.set_title("Время выполнения", fontsize=16)
ax.legend()
```

```python
fig = plt.figure(figsize = (15, 7))
ax = plt.axes()

ax.set_alpha(0.1)
plt.plot(size[:], time_1CPU[0::]  / time_1CPU[:], '-o', label = "Кол-во ядер: 1")
plt.plot(size[1::], time_1CPU[1::] / time_2CPU[:], '-o', label = "Кол-во ядер: 2")
plt.plot(size[2::], time_1CPU[2::] / time_4CPU[:], '-o', label = "Кол-во ядер: 4")
plt.plot(size[3::], time_1CPU[3::] / time_8CPU[:], '-o', label = "Кол-во ядер: 8")
plt.plot(size[4::], time_1CPU[4::] / time_16CPU[:], '-o', label = "Кол-во ядер: 16")

ax.set_xscale('log', base=2)
ax.set_yscale('linear')
ax.set_xlabel("Размер матрицы")
ax.set_ylabel("Время выполнения")
ax.xaxis.set_major_locator(FixedLocator([2**x for x in range(15)]))
ax.set_title("Spped Up", fontsize=16)
ax.legend()
```
